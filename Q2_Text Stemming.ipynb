```python
import re
from nltk.stem import PorterStemmer, LancasterStemmer

# Load the text from Data_1.txt
with open('Assignment Data\Data_1.txt', 'r') as file:
    text = file.read()

# Function to perform stemming using Regular Expression
def regex_stemmer(text):
    # Use a simple regex pattern to match word boundaries
    pattern = re.compile(r'\b\w+\b')
    words = pattern.findall(text)
    # Apply stemming by keeping only the first 4 characters of each word
    stemmed_words = [word[:4] for word in words]
    return ' '.join(stemmed_words)

# Function to perform stemming using Porter Stemmer
def porter_stemmer(text):
    porter = PorterStemmer()
    words = text.split()
    stemmed_words = [porter.stem(word) for word in words]
    return ' '.join(stemmed_words)

# Function to perform stemming using Lancaster Stemmer
def lancaster_stemmer(text):
    lancaster = LancasterStemmer()
    words = text.split()
    stemmed_words = [lancaster.stem(word) for word in words]
    return ' '.join(stemmed_words)

# Demonstrate word stemming
regex_output = regex_stemmer(text)
porter_output = porter_stemmer(text)
lancaster_output = lancaster_stemmer(text)

# Report the output
print("Regular Expression Stemmer Output:")
print(regex_output)
print("\nPorter Stemmer Output:")
print(porter_output)
print("\nLancaster Stemmer Output:")
print(lancaster_output)
```
