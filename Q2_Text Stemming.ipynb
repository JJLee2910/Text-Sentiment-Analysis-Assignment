```python
# 2. Demonstrate word stemming using Regular Expression, Porter Stemmer and Lancaster Stemmer and report the output. (6 marks)

# Import necessary library
import re
from nltk.stem import PorterStemmer, LancasterStemmer

class StemmingDemo:
    def __init__(self, text):
        self.text = text

    def regex_stemmer(self):
        pattern = re.compile(r'\b\w+\b')
        words = pattern.findall(self.text)
        stemmed_words = [word[:4] for word in words]
        return ' '.join(stemmed_words)

    def porter_stemmer(self):
        porter = PorterStemmer()
        words = self.text.split()
        stemmed_words = [porter.stem(word) for word in words]
        return ' '.join(stemmed_words)

    def lancaster_stemmer(self):
        lancaster = LancasterStemmer()
        words = self.text.split()
        stemmed_words = [lancaster.stem(word) for word in words]
        return ' '.join(stemmed_words)

if __name__ == "__main__":
    # Load the text from Data_1.txt
    with open('Assignment Data\Data_1.txt', 'r') as file:
        text_data = file.read()

    # Create an instance of the StemmingDemo class
    stemming_demo = StemmingDemo(text_data)

    # Demonstrate word stemming
    regex_output = stemming_demo.regex_stemmer()
    porter_output = stemming_demo.porter_stemmer()
    lancaster_output = stemming_demo.lancaster_stemmer()

    # Report the output
    print("Regular Expression Stemmer Output:")
    print(regex_output)
    print("\nPorter Stemmer Output:")
    print(porter_output)
    print("\nLancaster Stemmer Output:")
    print(lancaster_output)
```
