1.	Explain the importance of stemming in text analytics (4 marks)

-	Stemming is a primitive algorithmic process used in Natural Language Processing (NLP) to removes ending of words, stripping away prefixes and suffixes and transform the words to their base form. The main objective of stemming is to simplify and standardize words, which helps to improve the performance of information retrieval, text classification, and other tasks.

-	For example, the variations of word “run”, “runs”, “running”, and “ran” are treated as the same. By converting the words to their stems, NLP models can identify different forms of the same words as one entity, which can reduce the complexity of the text data. 

-	Moreover, stemming can reduce the dimensionality as it decreases the overall number of unique words in a dataset by merging variations into a single basic form. It is very important for text analysis tasks such as document classification and clustering, because it can enhance the efficiency of algorithms and decrease computational complexity.

-	 Next, Stemming can strengthen the search experience by finding similar documents even with small variations in the query terms. For instance, users might use the exact root form words, but stemming helps in getting related results despite linguistic differences. 

-	Last, stemming supports in uncovering patterns and sentiments by interpreting words with same meaning as equivalent in tasks like sentiment analysis and text mining. This leads to the overall accuracy of sentiment classification and the extraction of meaningful insights from large volumes of text data. 


```python
# 2. Demonstrate word stemming using Regular Expression, Porter Stemmer and Lancaster Stemmer and report the output. (6 marks)

import re
from nltk.stem import PorterStemmer, LancasterStemmer
from nltk.tokenize import word_tokenize

#Load the data
with open("Assignment Data\\Data_1.txt", "r") as file:
    data = file.read()

tokens = re.findall(r'\b\w+\b', data)

def stem(word):
    regexp = r'^(.*?)(ing|ly|ed|ious|ies|ive|es|s|ment)?$'
    stem, suffix = re.findall(regexp, word)[0]
    return stem

porter = PorterStemmer()
lancaster = LancasterStemmer()

print ("RE Stemmer output : ")
print([stem(t) for t in tokens])
print("\n")

print ("Porter Stemmer output : ")
print([porter.stem(words) for words in tokens])
print("\n")

print ("Lancaster Stemmer output : ")
print([lancaster.stem(words) for words in tokens])
```

The Regular Expression stemmer reduces words by using basic rules based on suffixes or patterns, yet it frequently lacks language accuracy.
